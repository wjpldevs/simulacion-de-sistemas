# Helper: Semilla segÃºn la fecha


```python
from datetime import datetime

def obtener_semilla_fecha():
    fecha = datetime.now()
    semilla_fecha = int(str(fecha)[len(str(fecha)) - 5:len(str(fecha)) - 1])
    return semilla_fecha*1234
```

# Helper: Obtener digitos medios

```python
def obtener_digitos_medios(x0,y0):
    longitud_x0 = len(str(x0))
    longitud_y0 = len(str(y0))
    array = list(map(int, str(y0)))
    array_medio = array[round(longitud_y0/longitud_x0) : longitud_x0 + round(longitud_y0/longitud_x0)]
    numero_medio = int(''.join(map(str, array_medio)))
    return numero_medio
```

# Helper: Mostrar tabla

```python
import matplotlib.pyplot as plt

def mostrar_tabla(data):
    figura, eje = plt.subplots(1,1)
    etiquetas_columna = ["Y", "X", "R"]
    eje.axis("tight")
    eje.axis("off")
    eje.table(cellText = data, colLabels = etiquetas_columna, loc = "center")
    plt.show()
```

# Algoritmo de cuadrados medios

```python
def algoritmo_cuadrados_medios(semilla, iteraciones):
    x0 = semilla
    longitud_x0 = len(str(x0))
    data = []
    if longitud_x0 > 3:
        y0 = x0**2
        x1 = obtener_digitos_medios(x0,y0)
        r1 = x1 / 10**longitud_x0
        x = x1
        r = r1
        row = [y0, x1, r1]
        data.append(row)
        for i in range(1,iteraciones):
            y =  x**2
            x = obtener_digitos_medios(x,y)
            r = x / 10**longitud_x0
            row = [y, x, r]
            data.append(row)
        return data
    else:
        return data
```

# Programa principal

```python
semilla = obtener_semilla_fecha()
iteraciones = 30
numeros_aleatorios = algoritmo_cuadrados_medios(semilla, iteraciones)
mostrar_tabla(numeros_aleatorios)
```
